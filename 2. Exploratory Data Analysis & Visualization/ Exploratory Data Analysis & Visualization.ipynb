# Exploratory Data Analysis

## Module 0: Introduction to Exploratory Data Analysis (EDA)

Before building machine learning models, it is critical to understand the structure, quality, and patterns within the data.

Exploratory Data Analysis (EDA) is the process of analyzing datasets to summarize their main characteristics, often using statistical methods and visualizations.

EDA helps answer key questions:

- What does the data look like?
- Are there missing or inconsistent values?
- What are the distributions of key variables?
- Are there relationships between features?
- Are there outliers that could affect model performance?

At this stage, the goal is not to build models, but to develop intuition about the dataset.

Well-executed EDA can:
- Improve model performance  
- Prevent incorrect assumptions  
- Reveal hidden patterns and trends  
- Identify data quality issues early  

In practice, EDA sits between data collection and model development, forming a critical step in the machine learning pipeline.

In this chapter, we will use Python libraries such as pandas, matplotlib, and seaborn to explore and visualize data in a structured way.

**EDA is where intuition meets data. Here, it determines whether a model succeeds or fails before training even begins.**

```python
# Run this pre-processing cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Module 1: Data Inspection

Once a dataset is loaded, the first step in Exploratory Data Analysis (EDA) is to inspect its structure and contents.

Data inspection helps answer foundational questions:

- How many rows and columns are present?
- What data types are assigned to each column?
- Are there missing values?
- What do the first few records look like?

Understanding these elements ensures that the dataset is usable and helps guide further analysis.

### Key Inspection Methods

- `head()` — view the first few rows  
- `tail()` — view the last few rows  
- `info()` — summary of columns, data types, and non-null counts  
- `describe()` — statistical summary of numerical features  
- `shape` — dimensions of the dataset  

These tools provide a quick overview of the dataset and help identify potential data quality issues early.

Data inspection is a critical step before visualization and modeling, as it establishes a clear understanding of the dataset’s structure and limitations.

```python
# Example: Load dataset
df = pd.read_csv("data/data.csv")

# View first and last rows
print(df.head())
print(df.tail())

# Dataset structure
print(df.shape)

# Column information
print(df.info())

# Statistical summary
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

Data inspection is often overlooked, but it is where most data issues are discovered before they impact analysis or model performance.

## Module 2: Univariate Analysis

Univariate analysis focuses on examining one variable at a time to understand its distribution, central tendency, and variability.

This step helps identify:

- The shape of the data distribution (normal, skewed, uniform)
- The presence of outliers
- The range and spread of values
- Potential data quality issues

Understanding individual variables is essential before analyzing relationships between features or building models.

### Common Techniques

- Histograms — visualize the distribution of numerical variables  
- Boxplots — detect outliers and understand spread  
- Value counts — summarize categorical variables  

Univariate analysis provides the first layer of insight into the dataset and helps guide feature selection and transformation decisions.

```python
# Histogram
sns.histplot(df["Age"], bins=20)
plt.title("Distribution of Age")
```

```python
# Boxplot
sns.boxplot(x=df["Score"])
plt.title("Salary Distribution with Outliers")
plt.show()
```

```python
# Value counts
print(df["Department"].value_counts())

# Bar plot
sns.countplot(x="Department", data=df)
plt.title("Department Distribution")
plt.xticks(rotation=45)
plt.show()
```

Univariate analysis is where patterns begin to emerge — it reveals how each feature behaves before interactions are introduced.

## Module 3: Bivariate Analysis

Bivariate analysis examines the relationship between two variables to identify patterns, trends, and potential correlations.

This step helps answer key questions:

- Do two variables move together?
- Is there a positive or negative relationship?
- Are there differences between groups?
- Are there patterns that could influence model performance?

Understanding relationships between variables is critical for feature selection and model building.

### Common Techniques

- Scatter plots — visualize relationships between two numerical variables  
- Boxplots — compare distributions across categories  
- Grouped summaries — analyze aggregated statistics  

Bivariate analysis builds on univariate insights and introduces interactions between features, which are essential for predictive modeling.

```python
# Scatter plot: Age vs Score
sns.scatterplot(x="Age", y="Score", data=df)
plt.title("Age vs Score")
plt.show()
```

```python
# Score distribution by Department
sns.boxplot(x="Department", y="Score", data=df)
plt.title("Score by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average score by department
grouped = df.groupby("Department")["Score"].mean()
print(grouped)
```

Bivariate analysis is where data begins to tell a story — relationships between variables often reveal the strongest signals for prediction.

## Module 4: Correlation Analysis

Correlation analysis measures the strength and direction of relationships between numerical variables.

It helps answer questions such as:

- Which features are strongly related?
- Do variables move together positively or negatively?
- Are there redundant features that provide similar information?

Correlation values range from -1 to 1:

- **1** → Perfect positive correlation  
- **0** → No correlation  
- **-1** → Perfect negative correlation  

### Why It Matters

- Helps identify important features for modeling  
- Detects multicollinearity (highly correlated independent variables)  
- Supports feature selection and dimensionality reduction  

### Important Note

Correlation does not imply causation. A strong correlation between two variables does not mean one causes the other.

```python
# Compute correlation matrix (numerical columns only)
corr = df.corr(numeric_only=True)

print(corr)
```

```python
# Heatmap of correlations
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Correlation analysis transforms visual patterns into measurable relationships, making it easier to identify which features matter most in a model.

## Module 5: Data Cleaning in Context

After exploring the dataset, the next step is to clean and prepare the data for analysis and modeling.

Data cleaning ensures that the dataset is consistent, complete, and reliable.

Common data issues include:

- Missing values  
- Duplicate records  
- Inconsistent formatting  
- Outliers or incorrect values  

### Key Cleaning Techniques

- Handling missing values (drop or impute)  
- Removing duplicates  
- Standardizing data formats  
- Filtering invalid or extreme values  

Cleaning should be guided by insights from EDA rather than applied blindly.

### Why It Matters

- Improves model accuracy  
- Prevents errors during training  
- Ensures consistency across features  

Data cleaning transforms raw data into a usable dataset, making it a critical step in the machine learning pipeline.

```python
# Check missing values
print(df.isnull().sum())

# Drop rows with missing values
df_clean = df.dropna()

# OR fill missing values (example: numerical column)\
df["Score"] = df["Score"].fillna(df["Score"].mean())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Example: filter unrealistic values
df = df[df["Age"] > 0]
df = df[df["Score"] >= 0]
```

Data cleaning is where data quality is enforced — small issues at this stage can lead to significant errors in model performance if left unresolved.
