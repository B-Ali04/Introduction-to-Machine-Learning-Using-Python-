## Chapter 3: Core Machine Learning

This chapter introduces the core machine learning algorithms used for prediction and classification tasks.

These models form the foundation of many real-world systems and are widely used across industries.

The focus here is not just on implementation, but on understanding when and why each model is used.

### 1. Linear Regression

Linear regression is one of the simplest and most commonly used algorithms for predicting continuous values.

It models the relationship between input variables and a target variable using a straight line.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([50, 60, 70, 80, 90])

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[6]])
print("Prediction:", prediction)
```

### 2. Polynomial Regression

Polynomial regression extends linear regression by allowing curved relationships between variables.

It is useful when data does not follow a straight-line pattern.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

prediction = model.predict(poly.transform([[6]]))
print("Prediction:", prediction)
```

### 3. Logistic Regression

Logistic regression is used for classification problems.

Instead of predicting continuous values, it predicts probabilities and assigns data to categories.

```python
from sklearn.linear_model import LogisticRegression

# Binary classification example
X = np.array([[20], [25], [30], [35], [40]])
y = np.array([0, 0, 1, 1, 1])

model = LogisticRegression()
model.fit(X, y)

prediction = model.predict([[28]])
print("Prediction:", prediction)
```

### 4. Decision Trees

Decision trees split data into branches based on feature values.

They are easy to interpret and can capture non-linear relationships.

```python
from sklearn.tree import DecisionTreeClassifier

X = np.array([[20], [25], [30], [35], [40]])
y = np.array([0, 0, 1, 1, 1])

model = DecisionTreeClassifier()
model.fit(X, y)

prediction = model.predict([[28]])
print("Prediction:", prediction)
```

### 5. Support Vector Machines (SVM)

Support Vector Machines are powerful models used for both classification and regression.

They work by finding the optimal boundary (hyperplane) that separates data into classes.


```python
from sklearn.svm import SVC

X = np.array([[200], [250], [300], [350], [400]])
y = np.array([0, 0, 1, 1, 1])

model = SVC()
model.fit(X, y)

prediction = model.predict([[28]])
print("Prediction:", prediction)
```

### Key Takeaways

Different models solve different types of problems.

- Linear models are fast and simple
- Tree-based models afe well-suited to handle complexity  
- SVMs are powerful but require fine tuning  

There is no individual "best model" performance, the best option depends on the data and how well it is prepared.

### Reality Check

In practice, multiple models are tested and compared before choosing the best one for a task.

The goal is not to pick the most complex model, but the one that performs best given the data and intended outcomes.
