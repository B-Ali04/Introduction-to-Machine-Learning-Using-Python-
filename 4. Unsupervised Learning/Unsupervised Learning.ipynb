## Chapter 4: Unsupervised Learning

Unsupervised learning focuses on identifying patterns in data without predefined labels.

Unlike supervised learning, there is no target variable. The goal is to explore the structure of the data, group similar observations, and uncover hidden relationships.

These methods are commonly used for segmentation, anomaly detection, and exploratory analysis.

### 1. K-Means Clustering

K-Means is one of the most widely used clustering algorithms.

It works by dividing data into a fixed number of clusters (K), where each data point belongs to the cluster with the nearest mean.

The algorithm iteratively updates cluster centers until the grouping stabilizes.

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([
    [1, 2], [2, 3], [3, 3],
    [8, 7], [9, 8], [10, 7]
])

# Create the model
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(x)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

print("Labels:", labels)
print("Centers:", centers)
```

```python
# Visualization
plt.scatter(x[:, 0], x[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='X')
plt.title("K-Means Clustering")
plt.show()
```

### 2. Choosing the Number of Clusters (K)

One challenge with K-Means is selecting the appropriate number of clusters.

A common method is the **Elbow Method**, which evaluates how clustering performance improves as K increases.

```python
inertia = []

for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 6), inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

### 3. Hierarchical Clustering

Hierarchical clustering builds a tree-like structure of clusters.

Instead of choosing the number of clusters upfront, it creates a hierarchy of possible groupings.

There are two main approaches:

- Agglomerative (bottom-up)  
- Divisive (top-down)  

Agglomerative clustering is the most commonly used.

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=2)
labels = model.fit_predict(x)

print("Labels:", labels)
```

```python
# Visualization 
plt.scatter(x[:, 0], x[:, 1], c=labels)
plt.title("Hierarchical Clustering")
plt.show()
```

### Key Takeaways

Unsupervised learning is about discovering structure, not predicting outcomes.

- K-Means is fast and works well when clusters are clearly separated  
- Hierarchical clustering provides more flexibility and insight into relationships  

The challenge is that there is no “correct” answer. Instead, you are expected to interpret patterns rather than predict labels.
