## Chapter 5: Model Evaluation & Optimization

Building a model is only part of the process. Evaluating and improving that model is what determines whether it is actually useful.

This chapter focuses on how to measure model performance, avoid common pitfalls like overfitting, and improve results through systematic tuning.

The goal is not just to train models, but to understand how well they generalize to new, unseen data.

### 1. Train/Test Split

To evaluate a model properly, the dataset is divided into two parts:

- Training data: used to train the model  
- Testing data: used to evaluate performance  

This ensures the model is tested on data it has never seen before.

```python
from sklearn.model_selection import train_test_split
import numpy as np

x = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 1, 1, 1])

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

print("Train size:", len(x_train))
print("Test size:", len(x_test))
```

### 2. Cross-Validation

Cross-validation improves reliability by splitting the data multiple times and averaging the results.

Instead of relying on a single train/test split, the model is evaluated across multiple subsets of the data.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

scores = cross_val_score(model, x, y, cv=3)

print("Cross-validation scores:", scores)
print("Average score:", scores.mean())
```

### 3. Hyperparameter Tuning

Hyperparameters are settings that control how a model learns.

Unlike model parameters, they are not learned from data and must be set manually or optimized through search techniques.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"]
}

model = SVC()

grid = GridSearchCV(model, param_grid, cv=3)
grid.fit(x, y)

print("Best parameters:", grid.best_params_)
```

### 4. Performance Metrics

Different problems require different evaluation metrics.

Choosing the correct metric is critical for understanding model performance.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_true = [0, 0, 1, 1]
y_pred = [0, 1, 1, 1]

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))
```

```python
# Regression metrics
from sklearn.metrics import mean_squared_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse)
```

### Key Takeaways

A model that performs well on training data but poorly on new data is not useful for most purposes.

- Train/test splits prevent overfitting  
- Cross-validation improves reliability  
- Hyperparameter tuning improves performance  
- Metrics define what “good” actually means  

If you are not evaluating your model properly, you are not really doing yourself any favors in machine learning, you are just fitting data.

### Reality Check

In practice, model evaluation is often more important than model selection.

A simpler model with reliable performance is usually preferred over a complex model that is unstable or difficult to interpret.
